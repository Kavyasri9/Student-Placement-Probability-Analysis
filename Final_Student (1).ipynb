```python
x = raw[['Age', 'Gender',
       'class_Type', 'Duration', 'Assignments submitted', 'projects submitted',
       'Mock_Test', 'Attendance(in %)',
       'Final score(in %)', 'Course Name_AI &ML',
       'Course Name_Cloud Computing', 'Course Name_Cybersecurity',
       'Course Name_Data Analytics', 'Course Name_Data Science',
       'Course Name_Ethical Hacking', 'Course Name_Python Programming',
       'Course Name_Robotics', 'Course Name_Software Testing',
       'Course Name_Web Development',
       'Certification obtained_AWS Specialization',
       'Certification obtained_Harvard University Professional Certificate',
       'Certification obtained_IIT Madras Course Completion',
       'Certification obtained_IIT Madras Standard Course',
       'Certification obtained_NASSCOM Certification',
       'Certification obtained_No Certification']]
```

```python
y = raw[['Placement status']]
```

                                                         Train-Test split(Data Validation)

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=42)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

                                                              Model Defining

```python
# Sklearn module

from sklearn.tree import DecisionTreeClassifier
```

```python
# Model Define

dt = DecisionTreeClassifier()

# Hyperparams: criterion, max_depth
```

                                                    Model Training on Train Data

                                                     DECISION TREE MODEL

```python
dt.fit(xtrain, ytrain)  # dt is the trained object model
```

                                                                      Model Params

```python
dt.get_params()
```

                                                                Feature Importance

```python
for i, j in zip(dt.feature_names_in_, dt.feature_importances_):
    print(i, ":", round(j, 2), end=", ")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
y.value_counts()
```

OutPut Column is imbalanced column so we are considering the F1-score,Precision and Recall

```python
# predictions on xtest data

ypred = dt.predict(xtest)
```

```python
ytest.value_counts()
```

```python
# Confusion Matrix

cm = confusion_matrix(ytest, ypred, labels=dt.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dt.classes_)
disp.plot()
plt.figure(figsize=(3,3))
plt.show()
```

```python
# Test Scores from Confusion Matrix

print(classification_report(ytest, ypred))
```

```python
# AUC-ROC Score

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(dt, xtest, ytest)
```

```python
# Sklearn library

from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt, x, y, cv=3)

# cv represents the number of folds that the data is trained (should be >= 2)
```

```python
scores
```

```python
# Taking average of scores

round(np.mean(scores),2)
```

```python
from sklearn.metrics import accuracy_score

trainscore = accuracy_score(ytrain, dt.predict(xtrain))
testscore = accuracy_score(ytest, dt.predict(xtest))

print("Train  Acc Score - {}".format(round(trainscore,2)))
print("Test  Acc Score - {}".format(round(testscore,2)))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

def PredictResponse():
    print("KNN built on the below Input Data:")
    
    # Load data (for feature reference)
    userinpdata = pd.read_csv("C://Users//KAVYASRI//OneDrive//Documents//Performance.csv")
    display(userinpdata.head())
    
    print("======================= Enter Choice of Input Data ====================")
    
    # User Inputs
    print("\nGender:", userinpdata.Gender.unique())
    gen = input()
    
    print("\nCourse Name:", userinpdata['Course Name'].unique())
    cn = input()
    
    print("\nclass_Type:", userinpdata['class_Type'].unique())
    ct = input()
    
    print("\nDuration:", userinpdata.Duration.unique())
    dt = input()
    
    age = eval(input(f"Enter Age Value ({userinpdata.Age.min()}-{userinpdata.Age.max()}): "))
    
    print("\nCertification obtained:", userinpdata['Certification obtained'].unique())
    tot = input()
    
    fd = eval(input(f"Assignments submitted ({userinpdata['Assignments submitted'].min()}-{userinpdata['Assignments submitted'].max()}): "))
    
    obr = eval(input(f"Projects submitted ({userinpdata['projects submitted'].min()}-{userinpdata['projects submitted'].max()}): "))
    
    scr = eval(input(f"Mock_Test ({userinpdata['Mock_Test'].min()}-{userinpdata['Mock_Test'].max()}): "))
    
    bhr = eval(input(f"Attendance(in %) ({userinpdata['Attendance(in %)'].min()}-{userinpdata['Attendance(in %)'].max()}): "))
    
    clr = eval(input(f"Final score(in %) ({userinpdata['Final score(in %)'].min()}-{userinpdata['Final score(in %)'].max()}): "))
    
    # Create a DataFrame with user inputs
    data = {
        'Gender': [gen],
        'Course Name': [cn],
        'class_Type': [ct],
        'Duration': [dt],
        'Age': [age],
        'Certification obtained': [tot],
        'Assignments submitted': [fd],
        'projects submitted': [obr],
        'Mock_Test': [scr],
        'Attendance(in %)': [bhr],
        'Final score(in %)': [clr]
    }
    
    row = pd.DataFrame(data)
    
    print("\nGiven User Input Data:")
    display(row)

    ####### Data Pre-Processing #######
    
    # Label Encoding
    row['Gender'] = row['Gender'].replace({'Female': 0, 'Male': 1})
    
    # One-Hot Encoding using stored encoder
    ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    
    # Fit on full dataset
    ohe.fit(userinpdata[['Course Name', 'Certification obtained']])  
    encoded_features = ohe.transform(row[['Course Name', 'Certification obtained']])
    
    encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out(['Course Name', 'Certification obtained']))
    
    row = row.drop(['Course Name', 'Certification obtained'], axis=1)
    row = pd.concat([row, encoded_df], axis=1)
    
    # Label Encoding for class_Type
    label_encoder = LabelEncoder()
    label_encoder.fit(userinpdata['class_Type'])
    row["class_Type"] = label_encoder.transform(row["class_Type"])
    
    # Convert Duration to numerical
    row['Duration'] = row['Duration'].str.replace(" months", "").astype(float)

    # Ensure all training features are present in prediction input
    trained_features = best_grid_search.feature_names_in_  # Get expected feature names
    for feature in trained_features:
        if feature not in row.columns:
            row[feature] = 0  # Add missing features with default value
    
    # Ensure feature order matches
    row = row[trained_features]

    print("\n********** Prediction ***********\n")
    
    prob0 = round(best_grid_search.predict_proba(row)[0][0], 2)
    prob1 = round(best_grid_search.predict_proba(row)[0][1], 2)
    
    print(f"Predicted Probability: {best_grid_search.classes_[0]} - {prob0}, {best_grid_search.classes_[1]} - {prob1}\n")
    
    out = best_grid_search.predict(row)[0]
    print("Prediction:", out)
```
